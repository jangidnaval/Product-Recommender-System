```python
def fun1(x):
    x=x.replace('[','')
    x=x.replace(']','')
    x=x.replace("'",'')
    x=x.split(sep=',')

    return int(x[0])

def fun2(x):
    x=x.replace('[','')
    x=x.replace(']','')
    x=x.replace("'",'')
    x=x.split(sep=',')

    return int(x[1])
```

```python
df['helpful_numerator']=df.helpful.apply(lambda x: fun1(x))
df['helpful_denominator']=df.helpful.apply(lambda x: fun2(x))
df.drop('helpful',inplace=True,axis=1)
```

```python
count = df.groupby("asin", as_index=False).count()[['asin','overall']]
#print(count.head())
count.rename(columns={'overall':'count'},inplace=True)
print(count.head())

count.to_csv('../csv/product_frequency.csv',index=False,sep=',')
```

```python
mean = df.groupby("asin", as_index=False).mean()
mean.drop('unixReviewTime',inplace=True,axis=1)
print(mean.head())
```
